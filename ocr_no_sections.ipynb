```python
from PIL import Image
import pytesseract
import time

def make_ocr(PNG,dir):
    screenshot=dir+PNG
    img = Image.open(screenshot)
    pytesseract.pytesseract.tesseract_cmd = '/usr/local/bin/tesseract'
    text = pytesseract.image_to_string(img, lang='pan')

    with open("7.txt", "a", encoding="utf-8") as f:
        f.write(text)
```

```python
for i in range(500):
    PNG=f"agri-8(p)_merged-{i}.png"
    if(PNG in pngs):
        make_ocr(PNG,"result/")
```
